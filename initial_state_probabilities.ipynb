```python
#open the pr_data file in read mode
pr_data=open(r"C:\Users\SOURAV MAITY\Downloads\pr_data.txt",'r')
file_handler=pr_data.read()
```

```python
#extraxt the amino acid sequences and their corresponding secondary_structures from the file
import re
def extract_single_quoted_substring(file_handler):
    #make a list of all the single quoted substring of the file
    a=re.findall(r"'(.*?)'",file_handler)
    return a
list_of_substrings=extract_single_quoted_substring(file_handler)
print(list_of_substrings)
#it will return a list containing both the amino acid sequences and the sequence of their corresponding  protein secondary structures 
```

```python
#now i have checked the length of each amino acid sequence and their corresponding protein secondary structure and store them in separate list if their size are equal
sequence_list=[list_of_substrings[i] for i in range(0,len(list_of_substrings),2) if len(list_of_substrings[i])==len(list_of_substrings[i+1]) and list_of_substrings[i+1].isalpha()]
structure_list=[list_of_substrings[j] for j in range(1,len(list_of_substrings),2) if len(list_of_substrings[j])==len(list_of_substrings[j-1]) and list_of_substrings[j].isalpha()]
print(len(sequence_list))
print(len(structure_list))
```

```python
#so there are 93 accurate datas are there
#we will use 70 datas to train our model
training_sequence_list=[sequence_list[k] for k in range(0,70)]
testing_sequence_list=[sequence_list[seq] for seq in range(70,len(sequence_list))]
print(testing_sequence_list,len(testing_sequence_list))
# print(training_sequence_list)
```

```python
#i have also used 70 corresponding structure data for training
training_structure_list=[structure_list[m] for m in range(0,70)]
testing_structure_list=[structure_list[str] for str in range(70,len(structure_list))]
print(testing_structure_list,len(testing_structure_list))
print(training_structure_list)
```

```python
#here i am considering the secondary structures are hidden states and amino acids are observed states
#so in this traing tructure list first i have identified unique secondary structures
a=""
for b in training_structure_list:
    a+=b
hidden_states=list(dict.fromkeys(a).keys())
hidden_states="".join(hidden_states)
print(hidden_states)
```

```python
#now i have calculated the initial state probabilities of the hidden states
def initial_state_probabilities(hidden_states,training_structure_list):
    d={n:0 for n in hidden_states}
    for o in training_structure_list:
        p=o[0]
        d[p]+=1
    for q in d:
        d[q]/=70
        r=round(d[q],2)
        d[q]=r
    return d
d=initial_state_probabilities(hidden_states,training_structure_list)
print(d)
```

```python
#so finally i get the hidden states as a string and initial state probabilities as a list
hidn_state=hidden_states
in_prob=list(d.values())
#THE PARAMETERS WE GET
print(hidn_state)
print(in_prob)
print(training_sequence_list)
# print(training_structure_list)
```
